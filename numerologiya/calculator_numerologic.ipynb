```python
# Значение (число) имени
def calculate_name_value(your_name):
    table_alpha = { 1:['а','и','с','ъ'],
                    2:['б','й','т','ы'],
                    3:['в','к','у','ь'],
                    4:['г','л','ф','э'],
                    5:['д','м','х','ю'],
                    6:['е','н','ц','я'],
                    7:['ё','о','ч'],
                    8:['ж','п','ш'],
                    9:['з','р','щ']}
    your_name = your_name.lower()
    count = 0
    for symbol in your_name:
        for key in table_alpha:
            if symbol in table_alpha[key]:
                count += key
    count = str(count)
    
    def calculate_1_number(number):
        result = str(sum([int(i) for i in number]))
        if len(result) > 1: # Если длинна строки больше 1
            # Вызываем функцию рекурсивно в виде аргумента(результат)
            return calculate_1_number(result)
        # Возвращаем полученный результат
        return result
    
    if len(count) > 1:
        count = calculate_1_number(count)
    return count
count_name = calculate_name_value(input_name)
print(f'Число имени: {count_name}')
```

```python
# texnicheskiy rasklad 
def technical_layout(input_date):
    # Очищаем дату от нулей и точек
    input_date = input_date.replace('.', '').replace('0', '')
    # Создаём список чисел из очищеной даты
    pure_numbers_list = [int(i) for i in input_date]
    # Получаем первое доп.число тела
    body_number1 = str(sum(pure_numbers_list))
    # Получаем второе доп.число тела
    if body_number1[1] == '0':
        body_number2 = body_number1[0]
    else:
        body_number2 = str(sum([int(i) for i in body_number1]))
    # Вычисляем число тела
    body = body_number1 + body_number2
    # Получаем первое доп.число души
    soul_number1 = str(abs(int(body_number1) - pure_numbers_list[0]*2))
    # Получаем второе доп.число души
    if len(soul_number1) == 1:
        soul_number2 = '0' + soul_number1
    else:
        soul_number2 = str(sum([int(i) for i in soul_number1]))
    # Вычисляем число души
    soul = soul_number1 + soul_number2
    
    # Возвращаем очищеную дату, число тела и число души
    return (input_date, body, soul)
# Дата рождения без лишних значений и тех.расклад души и тела
total_numbers = technical_layout(input_date)
# Тех.расклад души и тела
body_soul = total_numbers[1:3]
body_soul
```

```python
# Составляем таблицу Пифагора
# Создаём таблицу Пифагора ввиде аргумента принимаем полученый ранее тех.расклад души и тела 
# плюс дата рождения без нулей и лишних знаков
def calculate_table_piphagor(total_numbers):
    # Распакуем дату и тех.расклад
    date, body, soul = total_numbers
    # Сложим дату и тех.расклад в одну строку
    total = date + body + soul
    # Задаём размер нашей таблицы
    number_list = '123456789'
    # Подгоравливаем будущую таблицу в виде словаря
    piphagor_square = dict()
    # Задаём цикл по номерам таблицы
    for i in number_list:
        # Если для номера таблицы есть совпадения в строке с тех.данными 
        if i in total: 
            # ключём будет номер ячейки а значением количество совпадений
            piphagor_square[i]=i*total.count(i)
        # Если совпадений нету
        elif i not in total: 
            # ключём будет номер ячейки а значением пустота
            piphagor_square[i] = ''
    # Возвращаем таблицу Пифагора ввиде словаря, 
    # ключ=номер ячейки, значение=количество совпадений
    return piphagor_square
# Словарь таблица Пифагора
dict_table_piphagor = calculate_table_piphagor(total_numbers)
# Заносим ввиде списка, значения ключей
list_table_piphagor = list(calculate_table_piphagor(total_numbers).values())
print(list_table_piphagor)
```

```python
# Считаем число зрелости души
# Высчитываем число реинкарнации (зрелость души) по дате рождения
def calculate_reincarnation(list_table_piphagor):
    # Общая строка для вычисления
    total_str = ''
    # Пройдёмся циклом по полученому списку таблицы
    for i in list_table_piphagor:
        if i != '-': # Если ячейка не содержит прочерк, добавляем её в общую строку  
            total_str += i
    # Счиаем количество элементов в строке
    count_birth = len(total_str)
    result = None
    if count_birth < 10:
        result = f'Зрелость души:\n{count_birth}'
    else:
        result = f'Зрелость души:\n{count_birth}'
    return result
reincarnation = calculate_reincarnation(list_table_piphagor)
display(reincarnation)
```

```python
# Вычисление Роковых годов
def calculate_fateful_year(life_code):
    import numpy as np
    result = '' # Результат
    arr = np.arange(84) # Создаём одномерный массив до 83 лет
    arr.shape=(14,6) # Разделяем на двумерный массив
    table_13x6 = arr.transpose() # Меняем вертикальное положение на горизонтальное
    if '0' in life_code: # Если в Жизненном коде есть 'О' ноль
        rock_year_list = [] # Говорим переменную для списка Роковых годов
        for i, value in enumerate(life_code): # пройдемся цикло по жизненному коду
            if value == '0': # Если в ЖизКоде встретился '0' ноль
                search_list = table_13x6[i].tolist() # Выделяем весь список чисел этого столбца
                for j in search_list: # Пройдёмся по элементам выбранного списка
                    sun = str(round(int(life_code) / j)) # Жизненный код делим на число из списка
                    sun = sun[2] # Берём третий элемент (первое число солнца)
                    if sun == '0': # Если Солнце '0' ноль
                        rock_year_list.append(str(j)) # Добавляем этот Роковой год в список
        if rock_year_list: # Если список не пуст
            #result = sorted(rock_year_list) # Сортируем и заносим в переменную
            result = ','.join(sorted(rock_year_list)) # Сортируем и заносим в переменную в виде строки
            return result # Выводим результат
    if not result:
        return ''
    else:
        return ''
    

fateful_year = calculate_fateful_year(life_code) # Заносим в переменную, нашу функцию вычисления Роковых годов
print(f'год Рока: {fateful_year}')
```
